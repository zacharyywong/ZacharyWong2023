```python
import pandas as pd
from pandasql import sqldf 
import numpy as np
pysqldf = lambda q: sqldf(q, globals())
```

```python
#global variables 
pathtoData = '/Users/zacharywong/github/zacharywong2023/GreenWave/Data/'
pathtoDeliverables = '/Users/zacharywong/github/zacharywong2023/GreenWave/Deliverables/'
```

```python
# import bicycle count data in pandas dataframe 
bicycleCountData= 'Bike_Volume_-_Manual_Counts.csv' 
df = pd.read_csv(pathtoData + bicycleCountData)
df.head()

# Split Location into two streets
#df[['Street1', 'Street2']] = df['LOCATION'].str.split(' & ', 1, expand=True)

# reorder columns and delete unused columns 
df['YR2012_PM'] = np.nan

cols = df.columns.tolist()
del(cols[0:6])
del(cols[-7:-1])
del(cols[-4])
del(cols[-4])
cols[0:6] + cols[-1:] + cols[7:]
del(cols[-1])
df = df[cols]
df
```

```python
#missionStreets = ["""Church St|Dolores St|Gurerrero St|Valencia St|Mission St|S Vann Nessa St|Folsom St|Harrison St|Bryant St|Potrero St|
 #               13 St|14 St|15 St|16 St|17 St|18 St|19 St|20 St|21 St|22 St|23 St|24 St|25 St|26 St"""]
sfDF = df[(df['YR2006_PM'] > -1) & (df['YR2016_PM'] > -1)] 
#sfDF = df.T.reset_index()
#sfDF.columns = ['Count_Year', 'Bicycle_Count']
#sfDF
sfDF.head()
```

```python
# Use SQL to clean data 
# Only keep rows with streets in Mission District
# Streets: 13th-26th, Cesar Chavez
# Streets: Church, Dolores, Gurerrero, Valencia, Mission, S Vann Nessa, Folsom, Harrison, Bryant, Potrero

# split location into two streets

q = """SELECT Sum(YR2006_PM) as YR_2006, Sum(YR2007_PM) as YR_2007, Sum(YR2008_PM) as YR_2008, Sum(YR2009_PM) as YR_2009, 
        Sum(YR2010_PM) as YR_2010, Sum(YR2011_PM) as YR_2011, Sum(YR2013_PM) as YR_2013, Sum(YR2014_PM) as YR_2014, 
        Sum(YR2015_PM) as YR_2015, Sum(YR2016_PM) as YR_2016, Sum(YR2017_PM) as YR_2017
        From sfDF
    """ 
df = pysqldf(q)
#df.rename(index={'0': 'Row_1'})
#df.index = ['Bicycle Count']
dfBC = df.T.reset_index()
dfBC.columns = ['Count_Year', 'Bicycle_Count']
dfBC
```

```python
bicycleCountData= 'Bicycle_Parking.csv' 
df = pd.read_csv(pathtoData + bicycleCountData)
df

# clean up table and only keep spaces and install_year
cols = df.columns.tolist()
del cols[0:6]
del cols[-12:]
del cols[1]
cols
df = df[cols]
df
```

```python
# use sql to group by year 
q = """ SELECT INSTALL_YR as Year, SUM(SPACES) as Spaces
        FROM df
        WHERE INSTALL_YR >= 2006 AND INSTALL_YR <= 2017
        GROUP BY INSTALL_YR
        ORDER BY INSTALL_YR
    """
parkSFDF = pysqldf(q)
parkSFDF
```

```python
df = parkSFDF.join(dfBC, how='outer').drop(['Count_Year'], axis = 1)
df = df[['Year', 'Bicycle_Count', 'Spaces']]
df.set_index(['Year'])
fileName = 'SFData.csv'
df.to_csv(pathtoDeliverables + fileName)
```
