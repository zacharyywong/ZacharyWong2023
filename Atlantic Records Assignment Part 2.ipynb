```python
import pandas as pd

#Read in Top20 Shazamed Songs by City and US CSV files 
dfCityTop20 = pd.read_csv("/Users/zacharywong/github/zacharywong2023/Top20byUSCity.csv")
dfUSTop20 = pd.read_csv("/Users/zacharywong/github/zacharywong2023/Top20inUS.csv")

dfCityTop20.head()
```

```python
# Set indexes on the City Chart dataframe to Song Name and Artist 
dfCityTop20 = dfCityTop20.set_index(['Song Name', 'Artist'])
dfCityTop20.head()
```

```python
# Set indexes on the USTop20 DataFrame to Song Name and Artist

dfUSTop20 = dfUSTop20.set_index(['Song Name', 'Artist'])
dfUSTop20.head()
```

```python
# Concatenate the two dataframes into 1 dataframe
df = pd.concat([dfUSTop20, dfCityTop20], axis=1)
df
```

```python
#Export final, merged dataframe as CSV
df.to_csv('/Users/zacharywong/github/zacharywong2023/Top20inUSandCity.csv')
```
